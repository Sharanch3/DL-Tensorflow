```python
import numpy as np
from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.linear_model import LinearRegression
```

```python
X, y= load_diabetes(return_X_y= True)
print(f"X shape: {X.shape}")
print(f"y shape: {y.shape}")
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.2, random_state=2)
```

```python
sklearn_reg = LinearRegression()
sklearn_reg.fit(X_train, y_train)
```

```python
y_pred = sklearn_reg.predict(X_test)

print(f"SKlearn R2 score: {r2_score(y_true= y_test, y_pred= y_pred): .2f}%")
```

```python
class GDRegressor:

    def __init__(self, learning_rate = 0.5, epochs = 3000):
        self.coef_ = None
        self.intercept_ = None
        self.learning_rate = learning_rate
        self.epochs = epochs
        self.loss_history = []

    def fit(self, X_train, y_train):

        # initialize parameters
        self.intercept_ = np.random.randn()
        self.coef_ = np.random.randn(X_train.shape[1])

        for i in range(self.epochs):

            # prediction
            y_hat = np.dot(X_train, self.coef_) + self.intercept_

            # gradients
            grad_intercept_ = -2 * np.mean(y_train - y_hat)
            grad_coef_ = -2 * np.dot(X_train.T, (y_train - y_hat)) / X_train.shape[0]

            # parameter updates
            self.intercept_ -= self.learning_rate * grad_intercept_
            self.coef_ -= self.learning_rate * grad_coef_

            # loss tracking
            mse = np.mean((y_train - y_hat) ** 2)
            self.loss_history.append(mse)

    def predict(self, X_test):
        return np.dot(X_test, self.coef_) + self.intercept_
```

```python
gdr = GDRegressor()
gdr.fit(X_train, y_train)
```

```python
gdr.loss_history
```

```python
y_pred = gdr.predict(X_test= X_test)

r2_score(y_test, y_pred)
```
